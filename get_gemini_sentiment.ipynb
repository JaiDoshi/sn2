```python
selected_companies = ["AAPL", "KO", "V", "XOM", "TSLA"]
tweet_dir = "./tweet/raw/"


def get_tweets(ticker, date_str):
        print(f"Getting tweets for {ticker} on {date_str}")
        tweets = []
        tweet_path = os.path.join(tweet_dir, ticker, date_str)

        if os.path.exists(tweet_path):
            with open(tweet_path) as f:
                lines = f.readlines()
                for line in lines:
                    tweet_obj = json.loads(line)
                    tweets.append(tweet_obj['text'])
        return tweets

apple_daily_tweets = {}
for date_str in sorted(os.listdir(f"{tweet_dir}{selected_companies[0]}")):
    apple_daily_tweets[date_str] = get_tweets(selected_companies[0], date_str)
print(len(apple_daily_tweets))
```

## Prep training data for gemini

```python
from datetime import datetime, timedelta

def get_next_date(date_str):
    """
    Get the next day's date as a string.

    :param date_str: Date string in 'YYYY-MM-DD' format.
    :return: Next day's date as a string in 'YYYY-MM-DD' format.
    """
    date = datetime.strptime(date_str, '%Y-%m-%d')
    next_day = date + timedelta(days=1)
    return next_day.strftime('%Y-%m-%d')

def prepare_training_data(tweet_data, price_changes, system_instruction, training_data = []):
    """
    Prepare training data for fine-tuning.

    :param tweet_data: Dictionary with date as key and list of tweets as value.
    :param price_changes: Dictionary with date as key and price change (+1, -1) as value.
    :param system_instruction: System instruction string.
    :return: List of dictionaries formatted for fine-tuning.
    """

    for date, tweets in tweet_data.items():
        # Combine all tweets for the date into a single text block
        tweets_text = " ".join(tweets)
        
        # Ensure the next day's price change exists
        next_day = get_next_date(date)  # Implement get_next_date to fetch the next day
        if next_day in price_changes:
            label = price_changes[next_day]

            # Construct text_input with the system instruction and tweets
            text_input = f"{system_instruction} {tweets_text}"
            
            # Add the formatted example to the training data
            training_data.append({
                'text_input': text_input,
                'output': str(label)  # Convert label to string as expected in fine-tuning
            })
        else:
            continue
    return training_data
```

```python
apple_prompt = "Forget all your previous instructions. Pretend you are a financial expert. You are a financial expert with stock recommendation experience adept at analysing the effect of tweets on stock price. Predict the next-day price movement of the stock for Apple Inc. (AAPL) based on the following tweets. Your prediction should be binary,\
 1 for positive price movement and 0 for negative price movement. Only give the prediction as your response, nothing else. Here are the tweets from today: "

coke_prompt = "Forget all your previous instructions. Pretend you are a financial expert. You are a financial expert with stock recommendation experience adept at analysing the effect of tweets on stock price. Predict the next-day price movement of the stock for The Coca-Cola Company (KO) based on the following tweets. Your prediction should be binary,\
 1 for positive price movement and 0 for negative price movement. Only give the prediction as your response, nothing else. Here are the tweets from today: "

visa_prompt = "Forget all your previous instructions. Pretend you are a financial expert. You are a financial expert with stock recommendation experience adept at analysing the effect of tweets on stock price. Predict the next-day price movement of the stock for Visa Inc. (V) based on the following tweets. Your prediction should be binary,\
 1 for positive price movement and 0 for negative price movement. Only give the prediction as your response, nothing else. Here are the tweets from today: "

exxon_prompt = "Forget all your previous instructions. Pretend you are a financial expert. You are a financial expert with stock recommendation experience adept at analysing the effect of tweets on stock price. Predict the next-day price movement of the stock for Exxon Mobil Corporation (XOM) based on the following tweets. Your prediction should be binary,\
 1 for positive price movement and 0 for negative price movement. Only give the prediction as your response, nothing else. Here are the tweets from today: "

tesla_prompt = "Forget all your previous instructions. Pretend you are a financial expert. You are a financial expert with stock recommendation experience adept at analysing the effect of tweets on stock price. Predict the next-day price movement of the stock for Tesla, Inc. (TSLA) based on the following tweets. Your prediction should be binary,\
 1 for positive price movement and 0 for negative price movement. Only give the prediction as your response, nothing else. Here are the tweets from today: "
 
training_data = prepare_training_data(apple_daily_tweets, apple_label_dict, apple_prompt, [])
print(len(training_data))
training_data = prepare_training_data(coca_daily_tweets, coke_label_dict, coke_prompt,  training_data)
print(len(training_data))
training_data = prepare_training_data(visa_daily_tweets, visa_label_dict, visa_prompt, training_data)
print(len(training_data))
training_data = prepare_training_data(exxon_daily_tweets, exxon_label_dict, exxon_prompt, training_data)
print(len(training_data))
training_data = prepare_training_data(tesla_daily_tweets, tesla_label_dict, tesla_prompt, training_data)
print(len(training_data))
```

```python
base_model = "models/gemini-1.5-flash-001-tuning"

operation = genai.create_tuned_model(
    # You can use a tuned model here too. Set `source_model="tunedModels/..."`
    display_name="tune_all",
    source_model=base_model,
    epoch_count=10,
    batch_size=10,
    learning_rate=0.001,
    training_data=training_data,
    temperature=1,
)

for status in operation.wait_bar():
    time.sleep(10)

result = operation.result()
print(result)
# # You can plot the loss curve with:
# snapshots = pd.DataFrame(result.tuning_task.snapshots)
# sns.lineplot(data=snapshots, x='epoch', y='mean_loss')
```

```python
check = pd.read_csv("gemini_sentiment_predictions_finetuned.csv")['date_of_tweets'].to_list()
results = pd.read_csv("gemini_sentiment_predictions_finetuned.csv")
for date, tweets in apple_daily_tweets.items():
    if date in check:
        print(f"Skipping {date}")
        continue
    ticker = "AAPL"
    current_tweet = " ".join(tweets)
    company_name = f"Apple Inc. ({ticker})"
    prompt = f"Predict the next-day price movement of the stock for {company_name} based on the following tweets. Your prediction should be binary, 1 for positive price movement and 0 for negative price movement. Only give the prediction as your response, nothing else. Here are the tweets from today: {current_tweet}\n"
    try:
        response = gen_model.generate_content([prompt])
    except:
        time.sleep(6)
        response = gen_model.generate_content([prompt])
    try:
        results = pd.concat([results, pd.DataFrame([{
            "ticker": ticker,
            "date_of_tweets": date,
            "prediction": response.text,
            "confidence": np.exp(response.candidates[0].avg_logprobs)
        }])], ignore_index=True)
    except:
        print(f"Response not received for {date}")
        continue
    results.to_csv("gemini_sentiment_predictions_finetuned.csv", index=False)
    time.sleep(4)
    print(date, response.text, np.exp(response.candidates[0].avg_logprobs))
```
